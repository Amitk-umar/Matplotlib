```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# we use distribution plots to visualize the distribution of quantitative data.

Lets discuss some plots that allow us to visualize the distribution of a data set. these plots are :
• Distplot/histplot                                                                                                
• Jointplot
• pairplot
• rugplot


```python
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

```python
#try this also by commentout

# df = sns.displot(ld, x = 'total_bill', y = 'tip',bins = 50) #and histogram is also same thing you can use displot as histplot

plt.subplot(1,2,1) #1 row and 2 columns and first plot
df_1 = sns.histplot(df['total_bill'],bins = 20, kde= True, color = 'green',alpha= 0.6) 
df_1

plt.subplot(1,2,2) #1 row and 2 columns and second plot
df_2 = sns.histplot(df['tip'],bins = 20, kde= True, color = 'green',alpha= 0.6) 
df_2
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind= 'scatter')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind= 'resid')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind= 'kde')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind= 'hex')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind= 'hist')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind= 'reg')
```

```python
sns.pairplot(df,hue='size',palette = 'rainbow')
```

```python
sns.rugplot(df, height=0.5)
```
