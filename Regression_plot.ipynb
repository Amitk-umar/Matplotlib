# LMPLOT

regresion plot is used in machine learning so we will not fully discuss what ,how and where it is used but we will see how it is created and what are the values we can put inside it

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips  = sns.load_dataset('tips')
tips
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips,hue = 'sex', fit_reg = True,palette = 'Set1',markers = ['o','v'],scatter_kws = {'s':50}) #this is laso called linear regression plot
```
