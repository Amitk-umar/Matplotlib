```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Matplotlib Basic

```python
x = np.linspace(5,15,12)
print("For X\n",x)
y = x**2
print("\nFor Y\n",y)
```

```python
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Ploting a graph')
```

# Subplots in matplotlib

```python
plt.subplot(2,2,1)
plt.plot(x,y)

plt.subplot(2,2,2)
plt.plot(y,x,'y')

plt.subplot(2,2,4)
plt.plot(x**2,x**2,'purple')

plt.subplot(2,2,3)
plt.plot(y**3,x**3,'green')
```

# Object oriented way
matplotlib allows the aspect ratio , dpi and figsize to the specified when the figure object is created. you can use the figsize and dpi keyword arguments.

figsize is tuple of the width and height of the figure in inches.
dpi is the number of dots per inch (pixel per inch).

```python
fig = plt.figure(figsize = (8,4),dpi = 200)

axis_1 = fig.add_axes([0.1,0.1,1,1])
axis_1.plot(x,y)

axis_2 = fig.add_axes([0.3,0.6,0.3,0.3])
axis_2.plot(x,y)
```
