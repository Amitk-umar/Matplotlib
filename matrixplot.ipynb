```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

# HEAT MAP

```python
flights = sns.load_dataset("flights")
flights
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tipscorr  = tips[['total_bill','tip','size']]
tipscorr
```

```python
tipscorr.corr().plot()
```

```python
sns.heatmap(tipscorr.corr(),annot=True,linewidths=.5,fmt='.2f',cmap='rainbow')
```

# Cluster map


```python
sns.clustermap(tipscorr.corr(), method='average', cmap='coolwarm', figsize=(6, 5))
```

# Pivot table heat map

```python
pvt_table = flights.pivot_table(index='month',columns='year',values='passengers',observed=False)
pvt_table
```

```python
sns.heatmap(pvt_table)
```
